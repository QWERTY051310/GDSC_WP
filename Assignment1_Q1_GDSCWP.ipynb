```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

# Part 1

```python
plt.hist(iris.data[:,0])
plt.title('Sepal Length (cm)')
plt.show()
```

```python
plt.hist(iris.data[:,1])
plt.title('Sepal Width (cm)')
plt.show()
```

```python
plt.hist(iris.data[:,2])
plt.title('Petal Length (cm)')
plt.show()
```

```python
plt.hist(iris.data[:,3])
plt.title('Petal Width (cm)')
plt.show()
```

```python
plt.scatter(iris.data[:,0], iris.data[:,2])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

# Part 2

```python
print("Mean Sepal Length: " + str(np.mean(iris.data[:,0])))
print("Median Sepal Length: " + str(np.median(iris.data[:,0])))
print("Standard Deviation of Sepal Length: " + str(np.std(iris.data[:,0])))
```

```python
print("Mean Sepal Width: " + str(np.mean(iris.data[:,1])))
print("Median Sepal Width: " + str(np.median(iris.data[:,1])))
print("Standard Deviation of Sepal Width: " + str(np.std(iris.data[:,1])))
```

```python
print("Mean Petal Length: " + str(np.mean(iris.data[:,2])))
print("Median Petal Length: " + str(np.median(iris.data[:,2])))
print("Standard Deviation of Petal Length: " + str(np.std(iris.data[:,2])))
```

```python
print("Mean Petal Width: " + str(np.mean(iris.data[:,3])))
print("Median Petal Width: " + str(np.median(iris.data[:,3])))
print("Standard Deviation of Petal Width: " + str(np.std(iris.data[:,3])))
```
